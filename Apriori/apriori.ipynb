```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("market_basket - market_basket.csv", header=None)
df.head()
```

```python
df.shape
```

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(df.values[i, j]) for j in range(0, 20)])

from apyori import apriori

rules = apriori(transactions,
                min_support=0.003,
                min_confidence=0.2,
                min_lift=3,
                min_length=2)


results = list(rules)
result = [list(results[i][0]) for i in range(len(results))]
```
